```python
import os
import pygame

pygame.init()
```

```python
input_file = r"C:\Users\sergi\Downloads\MiniWorldSprites\MiniWorldSprites\Objects\FireballProjectile.png"
out_folder = r'C:\Users\sergi\Documents\repos\pygame_multi\assets\extracted'
name = 'FireballProjectile'
tile_size = 16

final_folder = os.path.join(out_folder, name)
os.makedirs(final_folder, exist_ok = True)

image = pygame.image.load(input_file)
width, height = image.get_rect().width, image.get_rect().height

for i in range(0, width, tile_size):
    for j in range(0, height, tile_size):
        clip_image = image.subsurface((i, j, tile_size, tile_size))
        pygame.image.save(clip_image, os.path.join(final_folder, f'{name}_{i}_{j}.png'))
```
